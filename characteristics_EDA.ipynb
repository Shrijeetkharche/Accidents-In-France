### Renaming columns

```python
# Column names
parameters = {
    'Num_Acc' : 'accident_id',
    'an'    : 'year',
    'mois'  : 'month',
    'jour'  : 'day',
    'hrmn'  : 'hour_min',
    'lum'   : 'lightning',
    'agg'   : 'localisation',
    'int'   : 'type_of_intersection',
    'atm'   : 'atmospheric_condition',
    'col'   : 'type_of_collision',
    'com'   : 'muncipality?',
    'adr'   : 'postal_address',
    'gps'   : 'gps_code',
    'lat'   : 'latitude',
    'long'  : 'longitude',
    'dep'   : 'department?'
}

# 


data_in = pd.read_csv('Dataset/caracteristics.csv', index_col=False, dtype='unicode', encoding='latin-1')
data_in.columns = [parameters[col] for col in data_in.columns]
data_in.head()

```

```python
temp = data_in.info()
```

### Checking for missing values

```python
df_temp = pd.DataFrame()

for col in data_in.columns:
    null_count = data_in[col].isnull().sum()
    total_count = null_count + data_in[col].count()
    percentage = round((null_count/total_count)*100, 2)
    df_temp = df_temp.append({'column': col, 'null_count': null_count, 'total_count': total_count, 'percentage': percentage}, ignore_index=True)
    # print(F"NULL- COUNT: {null_count}, TOTAL-COUNT: {total_count}, PERCENTAGE: {percentage}")
    # print(f'Percentage of missing values in {col}: {percentage}')

display(df_temp)
```
